```python
def importDataFromCSV(path, separator = ';'):
  return pd.read_csv(path, sep = separator)
```

```python
def analyseColumnEnds(df):
  copy = df.copy()
  columnNames = copy.columns.tolist()

  column_ranges = {}
  for column in copy.columns:
      if pd.api.types.is_numeric_dtype(copy[column]):
          column_ranges[column] = (copy[column].min(), copy[column].max())

  column_ranges = {}
  for column in copy.columns:
      if pd.api.types.is_numeric_dtype(copy[column]):
          column_ranges[column] = {
              "min": copy[column].min(),
              "max": copy[column].max()
          }

  return pd.DataFrame(column_ranges)
```

```python
def countClasses(df):
  copy = df.copy()
  ends = analyseColumnEnds(df)
  class_amount = 6
  result = pd.DataFrame()

  for column in copy.column():
    columnEnds = ends[column]
    step = (columnEnds['max'] - columnEnds['min'])/class_amount
    classes = [(columnEnds[min] + i * step) for i in range(class_amount)]

    counts = pd.Series()
    for cl in classes:
      count = copy[column].count(cl)


  return result
```

```python
def buildStatics(df):
  copy = df.copy()
  stats = pd.DataFrame({
        'Medidas': ['Média', 'Desvio padrão', 'Assimetria']
    })

  for column in copy.columns:
    stats[column] = [copy[column].mean(), copy[column].std(), copy[column].skew()]

  return stats
```

## Question two

```python
def filterQuality(df, low, up):
  filteredDataframe = df[(low <= df['quality']) & (df['quality'] < up)]

  stats = buildStatics(filteredDataframe)

  return stats
```
